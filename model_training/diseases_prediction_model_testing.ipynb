```python
def predict_disease():
    user_data = input("Input symptoms (comma-separated): ")
    symptoms_dict = load_symptoms()
    symptoms_list = [symptom.strip() for symptom in user_data.split(',')]
    symptoms_array = np.zeros(len(symptoms_dict))
    for symptom in symptoms_list:
        if symptom in symptoms_dict:
            symptoms_array[symptoms_dict[symptom]] = 1
    disease_index = model.predict(symptoms_array.reshape(1, -1))
    disease = lb_encoder.inverse_transform(disease_index)[0]
    return disease
```

```python
def predict_results():
    disease = predict_disease()
    medications, precautions, diets, workout = load_data()
    medi = medications[medications['Disease'] == disease]['Medication'].values[0]
    prec_row = precautions[precautions['Disease'] == disease]
    prec_list = [prec_row['Precaution_1'].values[0], prec_row['Precaution_2'].values[0],
                prec_row['Precaution_3'].values[0], prec_row['Precaution_4'].values[0]]
    prec = ", ".join([p for p in prec_list if pd.notna(p)])
    diet = diets[diets['Disease'] == disease]['Diet'].values[0]
    work = workout[workout['disease'] == disease]['workout'].values[0]
    return medi, prec, diet, work
```
